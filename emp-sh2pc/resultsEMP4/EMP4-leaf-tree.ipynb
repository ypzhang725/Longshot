```python
def string2vect(string, name):
   # print(string)
   # print(name)
    string_result = re.search(name + ': (.*)\n', string).group(1)
    vect = [float(e) for e in string_result.split(",")]

    return vect
```

```python
def plotDPSeperate13(T, arrayLeaf, arrayTree_sortD, xlable, ylable, title):
    xpoints = np.linspace(1, T, T, endpoint=True)
    x1points = np.linspace(1, 200, 200, endpoint=True)
    y1points = arrayLeaf
    y3points = arrayTree_sortD
    
    begin = y1points[0]
    end = y1points[-1]
    slope = (end - begin) / 200
    leaf = [(i-1)*slope+begin for i in xpoints]
    #plt.subplot(1, 1, 1)
    plt.plot(x1points, y1points, 'b^', label="Leaf")
    plt.plot(xpoints, leaf, 'c.', label="Leaf-linearline")
    plt.plot(xpoints, y3points, 'g^', label="Tree-optimized")

    plt.xlabel(xlable, fontsize=14)
    plt.ylabel(ylable, fontsize=14)
    plt.legend(loc="upper right", prop={'size': 14})
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.title(title, fontsize=14)
   # plt.yscale('log',base=10) 
    plt.savefig("plots200-1000/Q1-2-LeafTree-"+ylable+"-"+title+".pdf", bbox_inches='tight')

    plt.show()


    
def plotQ13(T, eps, N):
    # create file names
    T_string = str(T)+","
    eps_string = str(eps)+","
    N_string = str(N)
    fileNameLeaf = "leafbin40_ss1.txt,"+"200,"+eps_string+N_string+";2.txt"
    fileNameTreeSortD = "treebin40_ss1.txt,"+T_string+eps_string+N_string+",2;0.txt"

    # read files, parse and return mean
    returnedMeanLeaf = readFileReturnMeanAll(fileNameLeaf)
    returnedMeanSortD = readFileReturnMeanTree(fileNameTreeSortD)    
    
    # six metrics 
    title = "T:"+str(T)+"; eps:"+str(eps)+"; N:"+str(N)
    # point query
    # leaf 
    LeafDPError_mean = returnedMeanLeaf["metricDPErrorP_mean"]
    LeafDPStoreError_mean = returnedMeanLeaf["metricDPStoreErrorP_mean"]
    LeafTTStoreError_mean = returnedMeanLeaf["metricTTStoreErrorP_mean"]

    # tree_sortD
    SortDDPError_mean = returnedMeanSortD["metricDPErrorP_mean"]
    SortDDPStoreError_mean = returnedMeanSortD["metricDPStoreErrorP_mean"]
    SortDTTStoreError_mean = returnedMeanSortD["metricTTStoreErrorP_mean"]

    
    plotDPSeperate13(T, LeafDPError_mean, SortDDPError_mean, "The number of updates", "DPCountError-PointQuery", title)
    plotDPSeperate13(T, LeafTTStoreError_mean, SortDTTStoreError_mean, "The number of updates", "TrueRecordError-PointQuery", title)
    plotDPSeperate13(T, LeafDPStoreError_mean, SortDDPStoreError_mean, "The number of updates", "DPStoreError-PointQuery", title)
 
    # range query
    # leaf 
    LeafDPErrorRange_mean = returnedMeanLeaf["metricDPErrorR_mean"]
    LeafDPStoreErrorRange_mean = returnedMeanLeaf["metricDPStoreErrorR_mean"]
    LeafTTStoreErrorRange_mean = returnedMeanLeaf["metricTTStoreErrorR_mean"]
  
    # tree_sortD
    SortDDPErrorRange_mean = returnedMeanSortD["metricDPErrorR_mean"]
    SortDDPStoreErrorRange_mean = returnedMeanSortD["metricDPStoreErrorR_mean"]
    SortDTTStoreErrorRange_mean = returnedMeanSortD["metricTTStoreErrorR_mean"]

    
    plotDPSeperate13(T, LeafDPErrorRange_mean, SortDDPErrorRange_mean, "The number of updates", "DPCountError-RangeQuery", title)
    plotDPSeperate13(T, LeafTTStoreErrorRange_mean, SortDTTStoreErrorRange_mean, "The number of updates", "TrueRecordError-RangeQuery", title)
    plotDPSeperate13(T, LeafDPStoreErrorRange_mean, SortDDPStoreErrorRange_mean, "The number of updates", "DPStoreError-RangeQuery", title)

   # plotDPSeperate13(T, LeafDPErrorRange_mean, [0 for i in range(10)], SortDDPErrorRange_mean, "The number of updates", "DPCountError for range query", title)
   # plotDPSeperate13(T, LeafTTStoreErrorRange_mean, [0 for i in range(10)], SortDTTStoreErrorRange_mean, "The number of updates", "TrueRecordError for range query", title)
   # plotDPSeperate13(T, LeafDPStoreErrorRange_mean, [0 for i in range(10)], SortDDPStoreErrorRange_mean, "The number of updates", "DPStoreError for range query", title)


```
