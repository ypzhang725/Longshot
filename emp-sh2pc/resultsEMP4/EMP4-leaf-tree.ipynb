```python
def string2vect(string, name):
   # print(string)
   # print(name)
    string_result = re.search(name + ': (.*)\n', string).group(1)
    vect = [float(e) for e in string_result.split(",")]

    return vect
```

```python
def readFileReturnMeanAll(fileName):
    with open(fileName) as f:
        lines = f.readlines()
    numPrint = 12
    runs = int(len(lines) / numPrint)
    '''
    print(runs)
    for i in range(12):
        print(lines[i])
    ''' 
    metricRunTimeDP = [None]*runs
    metricRunTimeDPSort = [None]*runs
    metricRunTimeDPMerge = [None]*runs
    metricRunTimePointQuery = [None]*runs
    metricRunTimeRangeQuery = [None]*runs
    metricDPErrorP = [None]*runs
    metricDPStoreErrorP = [None]*runs
    metricTTStoreErrorP = [None]*runs
    metricDPErrorR = [None]*runs
    metricDPStoreErrorR = [None]*runs
    metricTTStoreErrorR = [None]*runs
    
    metricRunTimeDPSort = [None]*runs
    metricDPError = [None]*runs
    metricDPStoreError = [None]*runs
    for i in range(runs):
        metricRunTimeDP[i] = string2vect(lines[(i*numPrint) + 1], 'metricRunTimeDP')
        metricRunTimeDPSort[i] = string2vect(lines[(i*numPrint) + 2], 'metricRunTimeDPSort')
        metricRunTimeDPMerge[i] = string2vect(lines[(i*numPrint) + 3], 'metricRunTimeDPMerge')
        metricRunTimePointQuery[i] = string2vect(lines[(i*numPrint) + 4], 'metricRunTimePointQuery')
        metricRunTimeRangeQuery[i] = string2vect(lines[(i*numPrint) + 5], 'metricRunTimeRangeQuery')
        metricDPErrorP[i] = string2vect(lines[(i*numPrint) +6], 'metricDPErrorP')
        metricDPStoreErrorP[i] = string2vect(lines[(i*numPrint) + 7], 'metricDPStoreErrorP')
        metricTTStoreErrorP[i] = string2vect(lines[(i*numPrint) + 8], 'metricTTStoreErrorP')
        metricDPErrorR[i] = string2vect(lines[(i*numPrint) + 9], 'metricDPErrorR')
        metricDPStoreErrorR[i] = string2vect(lines[(i*numPrint) + 10], 'metricDPStoreErrorR')
        metricTTStoreErrorR[i] = string2vect(lines[(i*numPrint) + 11], 'metricTTStoreErrorR')
       

    metricRunTimeDP_mean = np.mean(np.array(metricRunTimeDP), axis = 0)
    metricRunTimeDPSort_mean = np.mean(np.array(metricRunTimeDPSort), axis = 0)
    metricRunTimeDPMerge_mean = np.mean(np.array(metricRunTimeDPMerge), axis = 0)
    metricRunTimePointQuery_mean = np.mean(np.array(metricRunTimePointQuery), axis = 0)
    metricRunTimeRangeQuery_mean = np.mean(np.array(metricRunTimeRangeQuery), axis = 0)
    metricDPErrorP_mean = np.mean(np.array(metricDPErrorP), axis = 0)
    metricDPStoreErrorP_mean = np.mean(np.array(metricDPStoreErrorP), axis = 0)
    metricTTStoreErrorP_mean = np.mean(np.array(metricTTStoreErrorP), axis = 0)
    metricDPErrorR_mean = np.mean(np.array(metricDPErrorR), axis = 0)
    metricDPStoreErrorR_mean = np.mean(np.array(metricDPStoreErrorR), axis = 0)
    metricTTStoreErrorR_mean = np.mean(np.array(metricTTStoreErrorR), axis = 0)
    
    d = dict(); 
    d['metricRunTimeDP_mean'] = metricRunTimeDP_mean
    d['metricRunTimeDPSort_mean'] = metricRunTimeDPSort_mean
    d['metricRunTimeDPMerge_mean'] = metricRunTimeDPMerge_mean
    d['metricRunTimePointQuery_mean'] = metricRunTimePointQuery_mean
    d['metricRunTimeRangeQuery_mean'] = metricRunTimeRangeQuery_mean
    d['metricDPErrorP_mean'] = metricDPErrorP_mean
    d['metricDPStoreErrorP_mean'] = metricDPStoreErrorP_mean
    d['metricTTStoreErrorP_mean'] = metricTTStoreErrorP_mean
    d['metricDPErrorR_mean'] = metricDPErrorR_mean
    d['metricDPStoreErrorR_mean'] = metricDPStoreErrorR_mean
    d['metricTTStoreErrorR_mean'] = metricTTStoreErrorR_mean
    '''
    print(metricRunTimeDP)
    print(metricRunTimeDPSort)
    print(metricRunTimeDPMerge)
    print(metricRunTimePointQuery)
    print(metricRunTimeRangeQuery)
    print(metricDPErrorP)
    print(metricDPStoreErrorP)
    print(metricTTStoreErrorP)
    print(metricDPErrorR)
    print(metricDPStoreErrorR)
    print(metricTTStoreErrorR)
    '''
    return d

    


```

```python
def plotDPSeperate13(T, arrayLeaf, arrayTree_sortD, xlable, ylable, title):
    xpoints = np.linspace(1, T, T, endpoint=True)
    x1points = np.linspace(1, 200, 200, endpoint=True)
    y1points = arrayLeaf
    y3points = arrayTree_sortD

    #plt.subplot(1, 1, 1)
    plt.plot(x1points, y1points, 'b^', label="Leaf")
    plt.plot(xpoints, y3points, 'g^', label="Tree-optimized")

    plt.xlabel(xlable, fontsize=14)
    plt.ylabel(ylable, fontsize=14)
    plt.legend(loc="upper right", prop={'size': 14})
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.title(title, fontsize=14)
   # plt.yscale('log',base=10) 
    plt.savefig("plots200-1000/Q1-2-LeafTree-"+ylable+"-"+title+".pdf", bbox_inches='tight')

    plt.show()


    
def plotQ13(T, eps, N):
    # create file names
    T_string = str(T)+","
    eps_string = str(eps)+","
    N_string = str(N)
    fileNameLeaf = "leafbin40_ss1.txt,"+"200,"+eps_string+N_string+";2.txt"
    fileNameTreeSortD = "treebin40_ss1.txt,"+T_string+eps_string+N_string+",2;0.txt"

    # read files, parse and return mean
    returnedMeanLeaf = readFileReturnMeanAll(fileNameLeaf)
    returnedMeanSortD = readFileReturnMeanTree(fileNameTreeSortD)    
    
    # six metrics 
    title = "T:"+str(T)+"; eps:"+str(eps)+"; N:"+str(N)
    # point query
    # leaf 
    LeafDPError_mean = returnedMeanLeaf["metricDPErrorP_mean"]
    LeafDPStoreError_mean = returnedMeanLeaf["metricDPStoreErrorP_mean"]
    LeafTTStoreError_mean = returnedMeanLeaf["metricTTStoreErrorP_mean"]

    # tree_sortD
    SortDDPError_mean = returnedMeanSortD["metricDPErrorP_mean"]
    SortDDPStoreError_mean = returnedMeanSortD["metricDPStoreErrorP_mean"]
    SortDTTStoreError_mean = returnedMeanSortD["metricTTStoreErrorP_mean"]

    
    plotDPSeperate13(T, LeafDPError_mean, SortDDPError_mean, "The number of updates", "DPCountError-PointQuery", title)
    plotDPSeperate13(T, LeafTTStoreError_mean, SortDTTStoreError_mean, "The number of updates", "TrueRecordError-PointQuery", title)
    plotDPSeperate13(T, LeafDPStoreError_mean, SortDDPStoreError_mean, "The number of updates", "DPStoreError-PointQuery", title)
 
    # range query
    # leaf 
    LeafDPErrorRange_mean = returnedMeanLeaf["metricDPErrorR_mean"]
    LeafDPStoreErrorRange_mean = returnedMeanLeaf["metricDPStoreErrorR_mean"]
    LeafTTStoreErrorRange_mean = returnedMeanLeaf["metricTTStoreErrorR_mean"]
  
    # tree_sortD
    SortDDPErrorRange_mean = returnedMeanSortD["metricDPErrorR_mean"]
    SortDDPStoreErrorRange_mean = returnedMeanSortD["metricDPStoreErrorR_mean"]
    SortDTTStoreErrorRange_mean = returnedMeanSortD["metricTTStoreErrorR_mean"]

    
    plotDPSeperate13(T, LeafDPErrorRange_mean, SortDDPErrorRange_mean, "The number of updates", "DPCountError-RangeQuery", title)
    plotDPSeperate13(T, LeafTTStoreErrorRange_mean, SortDTTStoreErrorRange_mean, "The number of updates", "TrueRecordError-RangeQuery", title)
    plotDPSeperate13(T, LeafDPStoreErrorRange_mean, SortDDPStoreErrorRange_mean, "The number of updates", "DPStoreError-RangeQuery", title)

   # plotDPSeperate13(T, LeafDPErrorRange_mean, [0 for i in range(10)], SortDDPErrorRange_mean, "The number of updates", "DPCountError for range query", title)
   # plotDPSeperate13(T, LeafTTStoreErrorRange_mean, [0 for i in range(10)], SortDTTStoreErrorRange_mean, "The number of updates", "TrueRecordError for range query", title)
   # plotDPSeperate13(T, LeafDPStoreErrorRange_mean, [0 for i in range(10)], SortDDPStoreErrorRange_mean, "The number of updates", "DPStoreError for range query", title)


```
