tree_sortSubRoot: sort all nodes for the subroot that need to be updated 

tree_sortD: sort the last d records for each bin for the subroot that need to be updated 



CountDP: $\sum_{bin} |DP count - True count|$

StoreDP: $\sum_{bin} |DP count - real records|$

StoreTT: $\sum_{bin} |True count - real records|$

SortTime: sort time for each update


T: total updates; epsAll: total privacy budget; N: #records in each cache;
Leaf approach: eps = epsAll;   Tree approach: epsTree = epsAll / log2(T)
