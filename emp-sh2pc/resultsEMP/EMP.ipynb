```python
def string2vect(string, name):
   # print(string)
   # print(name)
    string_result = re.search(name + ': (.*)\n', string).group(1)
    vect = [float(e) for e in string_result.split(",")]

    return vect
```

```python
def readFileReturnMeanAll(fileName):
    with open(fileName) as f:
        lines = f.readlines()
    numPrint = 12
    runs = int(len(lines) / numPrint)
    '''
    print(runs)
    for i in range(12):
        print(lines[i])
    ''' 
    metricRunTimeDP = [None]*runs
    metricRunTimeDPSort = [None]*runs
    metricRunTimeDPMerge = [None]*runs
    metricRunTimePointQuery = [None]*runs
    metricRunTimeRangeQuery = [None]*runs
    metricDPErrorP = [None]*runs
    metricDPStoreErrorP = [None]*runs
    metricTTStoreErrorP = [None]*runs
    metricDPErrorR = [None]*runs
    metricDPStoreErrorR = [None]*runs
    metricTTStoreErrorR = [None]*runs
    
    metricRunTimeDPSort = [None]*runs
    metricDPError = [None]*runs
    metricDPStoreError = [None]*runs
    for i in range(runs):
        metricRunTimeDP[i] = string2vect(lines[(i*numPrint) + 1], 'metricRunTimeDP')
        metricRunTimeDPSort[i] = string2vect(lines[(i*numPrint) + 2], 'metricRunTimeDPSort')
        metricRunTimeDPMerge[i] = string2vect(lines[(i*numPrint) + 3], 'metricRunTimeDPMerge')
        metricRunTimePointQuery[i] = string2vect(lines[(i*numPrint) + 4], 'metricRunTimePointQuery')
        metricRunTimeRangeQuery[i] = string2vect(lines[(i*numPrint) + 5], 'metricRunTimeRangeQuery')
        metricDPErrorP[i] = string2vect(lines[(i*numPrint) +6], 'metricDPErrorP')
        metricDPStoreErrorP[i] = string2vect(lines[(i*numPrint) + 7], 'metricDPStoreErrorP')
        metricTTStoreErrorP[i] = string2vect(lines[(i*numPrint) + 8], 'metricTTStoreErrorP')
        metricDPErrorR[i] = string2vect(lines[(i*numPrint) + 9], 'metricDPErrorR')
        metricDPStoreErrorR[i] = string2vect(lines[(i*numPrint) + 10], 'metricDPStoreErrorR')
        metricTTStoreErrorR[i] = string2vect(lines[(i*numPrint) + 11], 'metricTTStoreErrorR')
       

    metricRunTimeDP_mean = np.mean(np.array(metricRunTimeDP), axis = 0)
    metricRunTimeDPSort_mean = np.mean(np.array(metricRunTimeDPSort), axis = 0)
    metricRunTimeDPMerge_mean = np.mean(np.array(metricRunTimeDPMerge), axis = 0)
    metricRunTimePointQuery_mean = np.mean(np.array(metricRunTimePointQuery), axis = 0)
    metricRunTimeRangeQuery_mean = np.mean(np.array(metricRunTimeRangeQuery), axis = 0)
    metricDPErrorP_mean = np.mean(np.array(metricDPErrorP), axis = 0)
    metricDPStoreErrorP_mean = np.mean(np.array(metricDPStoreErrorP), axis = 0)
    metricTTStoreErrorP_mean = np.mean(np.array(metricTTStoreErrorP), axis = 0)
    metricDPErrorR_mean = np.mean(np.array(metricDPErrorR), axis = 0)
    metricDPStoreErrorR_mean = np.mean(np.array(metricDPStoreErrorR), axis = 0)
    metricTTStoreErrorR_mean = np.mean(np.array(metricTTStoreErrorR), axis = 0)
    
    d = dict(); 
    d['metricRunTimeDP_mean'] = metricRunTimeDP_mean
    d['metricRunTimeDPSort_mean'] = metricRunTimeDPSort_mean
    d['metricRunTimeDPMerge_mean'] = metricRunTimeDPMerge_mean
    d['metricRunTimePointQuery_mean'] = metricRunTimePointQuery_mean
    d['metricRunTimeRangeQuery_mean'] = metricRunTimeRangeQuery_mean
    d['metricDPErrorP_mean'] = metricDPErrorP_mean
    d['metricDPStoreErrorP_mean'] = metricDPStoreErrorP_mean
    d['metricTTStoreErrorP_mean'] = metricTTStoreErrorP_mean
    d['metricDPErrorR_mean'] = metricDPErrorR_mean
    d['metricDPStoreErrorR_mean'] = metricDPStoreErrorR_mean
    d['metricTTStoreErrorR_mean'] = metricTTStoreErrorR_mean
    '''
    print(metricRunTimeDP)
    print(metricRunTimeDPSort)
    print(metricRunTimeDPMerge)
    print(metricRunTimePointQuery)
    print(metricRunTimeRangeQuery)
    print(metricDPErrorP)
    print(metricDPStoreErrorP)
    print(metricTTStoreErrorP)
    print(metricDPErrorR)
    print(metricDPStoreErrorR)
    print(metricTTStoreErrorR)
    '''
    return d

    


```

```python
def plotQuery(T, arrayLeaf, arrayTree_baseline, arrayTree_sortD, xlable, ylable, title):
    xpoints = np.linspace(1, T, T, endpoint=True)
    x1points = np.linspace(1, 10, 10, endpoint=True)
    y1points = arrayLeaf
    y2points = arrayTree_baseline
    y3points = arrayTree_sortD
    
    #plt.subplot(1, 1, 1)
    plt.plot(xpoints, y1points, 'b^', label="runTime_leaf")
    plt.plot(x1points, y2points, 'r^', label="runTime_baseline")
    plt.plot(xpoints, y3points, 'g^', label="runTime_tree_sortD")

    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.legend(loc="upper left", prop={'size': 8})
    plt.yscale('log',base=10) 
    
    plt.title(title)
    plt.savefig("plots100-1000/Q1-1-"+ylable+"-"+title+".png", bbox_inches='tight')
    plt.show()
```

```python
def plotDPSeperate(T, arrayLeaf, arrayTree_sortSubroot, arrayTree_sortD, xlable, ylable, title):
    xpoints = np.linspace(1, T, T, endpoint=True)
    x1points = np.linspace(1, 10, 10, endpoint=True)
    y1points = arrayLeaf
    y2points = arrayTree_sortSubroot
    y3points = arrayTree_sortD

    #plt.subplot(1, 1, 1)
    plt.plot(xpoints, y1points, 'b^', label="leaf")
    plt.plot(x1points, y2points, 'r^', label="baseline")
    plt.plot(xpoints, y3points, 'g^', label="tree_sortD")

    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.legend(loc="upper left", prop={'size': 8})

    plt.title(title)
    plt.savefig("plots100-1000/Q1-2-"+ylable+"-"+title+".png", bbox_inches='tight')

    plt.show()


    
def plotQ12(T, eps, N):
    # create file names
    T_string = str(T)+","
    eps_string = str(eps)+","
    N_string = str(N)
    fileNameLeaf = "leafbin40_ss1.txt,"+T_string+eps_string+N_string+";2.txt"
    fileNameBaseline = "baselinebin40_ss1.txt,"+"10,"+eps_string+N_string+";2.txt"
    fileNameTreeSortD = "treebin40_ss1.txt,"+T_string+eps_string+N_string+",2;0.txt"

    # read files, parse and return mean
    returnedMeanLeaf = readFileReturnMeanAll(fileNameLeaf)
    returnedMeanBaseline = readFileReturnMeanBaseline(fileNameBaseline)
    returnedMeanSortD = readFileReturnMeanTree(fileNameTreeSortD)    
    
    # six metrics 
    title = "T:"+str(T)+"; eps:"+str(eps)+"; N:"+str(N)
    # point query
    # leaf 
    LeafDPError_mean = returnedMeanLeaf["metricDPErrorP_mean"]
    LeafDPStoreError_mean = returnedMeanLeaf["metricDPStoreErrorP_mean"]
    LeafTTStoreError_mean = returnedMeanLeaf["metricTTStoreErrorP_mean"]
    # baseline
    BaselineDPError_mean = returnedMeanBaseline["metricDPErrorPoint_mean"]
    BaselineDPStoreError_mean = returnedMeanBaseline["metricDPStoreErrorPoint_mean"]
    BaselineTTStoreError_mean = returnedMeanBaseline["metricTTStoreErrorPoint_mean"]
    # tree_sortD
    SortDDPError_mean = returnedMeanSortD["metricDPErrorP_mean"]
    SortDDPStoreError_mean = returnedMeanSortD["metricDPStoreErrorP_mean"]
    SortDTTStoreError_mean = returnedMeanSortD["metricTTStoreErrorP_mean"]

    
    plotDPSeperate(T, LeafDPError_mean, BaselineDPError_mean, SortDDPError_mean, "time", "DPError-pointQ", title)
    plotDPSeperate(T, LeafTTStoreError_mean, BaselineTTStoreError_mean, SortDTTStoreError_mean, "time", "TrueStoreError-pointQ", title)
    plotDPSeperate(T, LeafDPStoreError_mean, BaselineDPStoreError_mean, SortDDPStoreError_mean, "time", "DPStoreError-pointQ", title)
 
    # range query
    # leaf 
    LeafDPErrorRange_mean = returnedMeanLeaf["metricDPErrorR_mean"]
    LeafDPStoreErrorRange_mean = returnedMeanLeaf["metricDPStoreErrorR_mean"]
    LeafTTStoreErrorRange_mean = returnedMeanLeaf["metricTTStoreErrorR_mean"]
    # baseline
    BaselineDPErrorRange_mean = returnedMeanBaseline["metricDPErrorRange_mean"]
    BaselineDPStoreErrorRange_mean = returnedMeanBaseline["metricDPStoreErrorRange_mean"]
    BaselineTTStoreErrorRange_mean = returnedMeanBaseline["metricTTStoreErrorRange_mean"]
    # tree_sortD
    SortDDPErrorRange_mean = returnedMeanSortD["metricDPErrorR_mean"]
    SortDDPStoreErrorRange_mean = returnedMeanSortD["metricDPStoreErrorR_mean"]
    SortDTTStoreErrorRange_mean = returnedMeanSortD["metricTTStoreErrorR_mean"]

    
    plotDPSeperate(T, LeafDPErrorRange_mean, BaselineDPErrorRange_mean, SortDDPErrorRange_mean, "time", "DPError-rangeQ", title)
    plotDPSeperate(T, LeafTTStoreErrorRange_mean, BaselineTTStoreErrorRange_mean, SortDTTStoreErrorRange_mean, "time", "TrueStoreError-rangeQ", title)
    plotDPSeperate(T, LeafDPStoreErrorRange_mean, BaselineDPStoreErrorRange_mean, SortDDPStoreErrorRange_mean, "time", "DPStoreError-rangeQ", title)


```

```python
def plotDP1Approach(T, eps1, eps2, eps3, xlable, ylable, title):
    xpoints = np.linspace(1, T, T, endpoint=True)
    y1points = eps1
    y2points = eps2
    y3points = eps3

    #plt.subplot(1, 1, 1)
    plt.plot(xpoints, y1points, 'b^', label="eps1")
    plt.plot(xpoints, y2points, 'r^', label="eps2")
    plt.plot(xpoints, y3points, 'g^', label="eps3")

    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.legend(loc="upper left", prop={'size': 8})

    plt.title(title)
    plt.savefig("plots100-1000/Q3-"+ylable+"-"+title+".png", bbox_inches='tight')
    plt.show()


    
def plotSort1Approach(T, eps1, eps2, eps3, xlable, ylable, title):
    xpoints = np.linspace(1, T, T, endpoint=True)
    y1points = eps1
    y2points = eps2
    y3points = eps3
    
    #plt.subplot(1, 1, 1)
    plt.plot(xpoints, y1points, 'b^', label="sortTime_eps1")
    plt.plot(xpoints, y2points, 'r^', label="sortTime_tree_eps2")
    plt.plot(xpoints, y3points, 'g^', label="sortTime_tree_eps3")

    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.legend(loc="upper left", prop={'size': 8})

    plt.title(title)
    plt.savefig("plots100-1000/Q3-"+ylable+"-"+title+".png", bbox_inches='tight')
    plt.show()
    
def plotQ31(T, epsList, N):
    # create file names
    T_string = str(T)+","
    N_string = str(N)
    LeafDPError_mean_list = []
    LeafDPStoreError_mean_list = []
    LeafTTStoreError_mean_list = []
    LeafDPErrorRange_mean_list = []
    LeafDPStoreErrorRange_mean_list = []
    LeafTTStoreErrorRange_mean_list = []
    
    SortDDPError_mean_list = []
    SortDDPStoreError_mean_list = []
    SortDTTStoreError_mean_list = []
    SortDDPErrorRange_mean_list = []
    SortDDPStoreErrorRange_mean_list = []
    SortDTTStoreErrorRange_mean_list = []
    
    LeafDRunTimeDPSort_mean_list = []
    SortDRunTimeDPSort_mean_list = []
    
    for eps in epsList:
        eps_string = str(eps)+","
        fileNameLeaf = "leafbin40_ss1.txt,"+T_string+eps_string+N_string+";2.txt"
        fileNameTreeSortD = "treebin40_ss1.txt,"+T_string+eps_string+N_string+",2;0.txt"
        # read files, parse and return mean
        returnedMeanLeaf = readFileReturnMeanAll(fileNameLeaf)
        returnedMeanSortD = readFileReturnMeanTree(fileNameTreeSortD)    
        # point query
        # leaf 
        LeafDPError_mean = returnedMeanLeaf["metricDPErrorP_mean"]
        LeafDPStoreError_mean = returnedMeanLeaf["metricDPStoreErrorP_mean"]
        LeafTTStoreError_mean = returnedMeanLeaf["metricTTStoreErrorP_mean"]
        
        LeafDPError_mean_list.append(LeafDPError_mean)
        LeafDPStoreError_mean_list.append(LeafDPStoreError_mean)
        LeafTTStoreError_mean_list.append(LeafTTStoreError_mean)
        
        # tree_sortD
        SortDDPError_mean = returnedMeanSortD["metricDPErrorP_mean"]
        SortDDPStoreError_mean = returnedMeanSortD["metricDPStoreErrorP_mean"]
        SortDTTStoreError_mean = returnedMeanSortD["metricTTStoreErrorP_mean"]
        
        SortDDPError_mean_list.append(SortDDPError_mean)
        SortDDPStoreError_mean_list.append(SortDDPStoreError_mean)
        SortDTTStoreError_mean_list.append(SortDTTStoreError_mean)
        
        
        # range query
        # leaf 
        LeafDPErrorRange_mean = returnedMeanLeaf["metricDPErrorR_mean"]
        LeafDPStoreErrorRange_mean = returnedMeanLeaf["metricDPStoreErrorR_mean"]
        LeafTTStoreErrorRange_mean = returnedMeanLeaf["metricTTStoreErrorR_mean"]
        
        LeafDPErrorRange_mean_list.append(LeafDPErrorRange_mean)
        LeafDPStoreErrorRange_mean_list.append(LeafDPStoreErrorRange_mean)
        LeafTTStoreErrorRange_mean_list.append(LeafTTStoreErrorRange_mean)
        
        # tree_sortD
        SortDDPErrorRange_mean = returnedMeanSortD["metricDPErrorR_mean"]
        SortDDPStoreErrorRange_mean = returnedMeanSortD["metricDPStoreErrorR_mean"]
        SortDTTStoreErrorRange_mean = returnedMeanSortD["metricTTStoreErrorR_mean"]
        
        SortDDPErrorRange_mean_list.append(SortDDPErrorRange_mean)
        SortDDPStoreErrorRange_mean_list.append(SortDDPStoreErrorRange_mean)
        SortDTTStoreErrorRange_mean_list.append(SortDTTStoreErrorRange_mean)
        
        
        LeafDRunTimeDPSort_mean = returnedMeanLeaf["metricRunTimeDPSort_mean"]
        SortDRunTimeDPSort_mean = returnedMeanSortD["metricRunTimeDPSort_mean"]
        LeafDRunTimeDPSort_mean_list.append(LeafDRunTimeDPSort_mean)
        SortDRunTimeDPSort_mean_list.append(SortDRunTimeDPSort_mean)

        
    # six metrics 
   # title = "T:"+str(T)+"; eps:"+str(eps)+"; N:"+str(N)
    
    title = "Q3"
    
    plotDP1Approach(T, LeafDPError_mean_list[0], LeafDPError_mean_list[1], LeafDPError_mean_list[2], "time", "LeafDPError", title)
    plotDP1Approach(T, LeafDPStoreError_mean_list[0], LeafDPStoreError_mean_list[1], LeafDPStoreError_mean_list[2], "time", "LeafDPStoreError", title)
    plotDP1Approach(T, LeafTTStoreError_mean_list[0], LeafTTStoreError_mean_list[1], LeafTTStoreError_mean_list[2], "time", "LeafTTStoreError", title)
 
    plotDP1Approach(T, LeafDPErrorRange_mean_list[0], LeafDPErrorRange_mean_list[1], LeafDPErrorRange_mean_list[2], "time", "LeafDPErrorRange", title)
    plotDP1Approach(T, LeafDPStoreErrorRange_mean_list[0], LeafDPStoreErrorRange_mean_list[1], LeafDPStoreErrorRange_mean_list[2], "time", "LeafDPStoreErrorRange", title)
    plotDP1Approach(T, LeafTTStoreErrorRange_mean_list[0], LeafTTStoreErrorRange_mean_list[1], LeafTTStoreErrorRange_mean_list[2], "time", "LeafTTStoreErrorRange", title)

    plotDP1Approach(T, SortDDPError_mean_list[0], SortDDPError_mean_list[1], SortDDPError_mean_list[2], "time", "SortDDPError", title)
    plotDP1Approach(T, SortDDPStoreError_mean_list[0], SortDDPStoreError_mean_list[1], SortDDPStoreError_mean_list[2], "time", "SortDDPStoreError", title)
    plotDP1Approach(T, SortDTTStoreError_mean_list[0], SortDTTStoreError_mean_list[1], SortDTTStoreError_mean_list[2], "time", "SortDTTStoreError", title)
 
    plotDP1Approach(T, SortDDPErrorRange_mean_list[0], SortDDPErrorRange_mean_list[1], SortDDPErrorRange_mean_list[2], "time", "SortDDPErrorRange", title)
    plotDP1Approach(T, SortDDPStoreErrorRange_mean_list[0], SortDDPStoreErrorRange_mean_list[1], SortDDPStoreErrorRange_mean_list[2], "time", "SortDDPStoreErrorRange", title)
    plotDP1Approach(T, SortDTTStoreErrorRange_mean_list[0], SortDTTStoreErrorRange_mean_list[1], SortDTTStoreErrorRange_mean_list[2], "time", "SortDTTStoreErrorRange", title)


    plotSort(T, LeafDRunTimeDPSort_mean_list[0], LeafDRunTimeDPSort_mean_list[1], LeafDRunTimeDPSort_mean_list[2], "time", "Updating DPStore(s) for leaf ", title)
    plotSort(T, SortDRunTimeDPSort_mean_list[0], SortDRunTimeDPSort_mean_list[1], SortDRunTimeDPSort_mean_list[2], "time", "Updating DPStore(s) for tree", title)



    
    
    
```
