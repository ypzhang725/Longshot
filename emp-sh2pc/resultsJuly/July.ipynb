```python
def string2vect(string, name):
    string_result = re.search(name + ': (.*)\n', string).group(1)
    vect = [float(e) for e in string_result.split(",")]

    return vect
```

```python
def readFileReturnMeanAll(fileName):
    with open(fileName) as f:
        lines = f.readlines()
    numPrint = 12
    runs = int(len(lines) / numPrint)
    '''
    print(runs)
    for i in range(12):
        print(lines[i])
    '''    
    metricRunTimeDP = [None]*runs
    metricRunTimeDPSort = [None]*runs
    metricRunTimeDPMerge = [None]*runs
    metricRunTimePointQuery = [None]*runs
    metricRunTimeRangeQuery = [None]*runs
    metricDPErrorP = [None]*runs
    metricDPStoreErrorP = [None]*runs
    metricTTStoreErrorP = [None]*runs
    metricDPErrorR = [None]*runs
    metricDPStoreErrorR = [None]*runs
    metricTTStoreErrorR = [None]*runs
    
    metricRunTimeDPSort = [None]*runs
    metricDPError = [None]*runs
    metricDPStoreError = [None]*runs
    for i in range(runs):
        metricRunTimeDP[i] = string2vect(lines[(i*numPrint) + 1], 'metricRunTimeDP')
        metricRunTimeDPSort[i] = string2vect(lines[(i*numPrint) + 2], 'metricRunTimeDPSort')
        metricRunTimeDPMerge[i] = string2vect(lines[(i*numPrint) + 3], 'metricRunTimeDPMerge')
        metricRunTimePointQuery[i] = string2vect(lines[(i*numPrint) + 4], 'metricRunTimePointQuery')
        metricRunTimeRangeQuery[i] = string2vect(lines[(i*numPrint) + 5], 'metricRunTimeRangeQuery')
        metricDPErrorP[i] = string2vect(lines[(i*numPrint) +6], 'metricDPErrorP')
        metricDPStoreErrorP[i] = string2vect(lines[(i*numPrint) + 7], 'metricDPStoreErrorP')
        metricTTStoreErrorP[i] = string2vect(lines[(i*numPrint) + 8], 'metricTTStoreErrorP')
        metricDPErrorR[i] = string2vect(lines[(i*numPrint) + 9], 'metricDPErrorR')
        metricDPStoreErrorR[i] = string2vect(lines[(i*numPrint) + 10], 'metricDPStoreErrorR')
        metricTTStoreErrorR[i] = string2vect(lines[(i*numPrint) + 11], 'metricTTStoreErrorR')
       

    metricRunTimeDP_mean = np.mean(np.array(metricRunTimeDP), axis = 0)
    metricRunTimeDPSort_mean = np.mean(np.array(metricRunTimeDPSort), axis = 0)
    metricRunTimeDPMerge_mean = np.mean(np.array(metricRunTimeDPMerge), axis = 0)
    metricRunTimePointQuery_mean = np.mean(np.array(metricRunTimePointQuery), axis = 0)
    metricRunTimeRangeQuery_mean = np.mean(np.array(metricRunTimeRangeQuery), axis = 0)
    metricDPErrorP_mean = np.mean(np.array(metricDPErrorP), axis = 0)
    metricDPStoreErrorP_mean = np.mean(np.array(metricDPStoreErrorP), axis = 0)
    metricTTStoreErrorP_mean = np.mean(np.array(metricTTStoreErrorP), axis = 0)
    metricDPErrorR_mean = np.mean(np.array(metricDPErrorR), axis = 0)
    metricDPStoreErrorR_mean = np.mean(np.array(metricDPStoreErrorR), axis = 0)
    metricTTStoreErrorR_mean = np.mean(np.array(metricTTStoreErrorR), axis = 0)
    
    d = dict(); 
    d['metricRunTimeDP_mean'] = metricRunTimeDP_mean
    d['metricRunTimeDPSort_mean'] = metricRunTimeDPSort_mean
    d['metricRunTimeDPMerge_mean'] = metricRunTimeDPMerge_mean
    d['metricRunTimePointQuery_mean'] = metricRunTimePointQuery_mean
    d['metricRunTimeRangeQuery_mean'] = metricRunTimeRangeQuery_mean
    d['metricDPErrorP_mean'] = metricDPErrorP_mean
    d['metricDPStoreErrorP_mean'] = metricDPStoreErrorP_mean
    d['metricTTStoreErrorP_mean'] = metricTTStoreErrorP_mean
    d['metricDPErrorR_mean'] = metricDPErrorR_mean
    d['metricDPStoreErrorR_mean'] = metricDPStoreErrorR_mean
    d['metricTTStoreErrorR_mean'] = metricTTStoreErrorR_mean
    '''
    print(metricRunTimeDP)
    print(metricRunTimeDPSort)
    print(metricRunTimeDPMerge)
    print(metricRunTimePointQuery)
    print(metricRunTimeRangeQuery)
    print(metricDPErrorP)
    print(metricDPStoreErrorP)
    print(metricTTStoreErrorP)
    print(metricDPErrorR)
    print(metricDPStoreErrorR)
    print(metricTTStoreErrorR)
    '''
    return d

    


```

```python
def readFileReturnMeanBaseline(fileName):
    with open(fileName) as f:
        lines = f.readlines()
    numPrint = 9
    runs = int(len(lines) / numPrint)
    '''
    print(runs)
    for i in range(9):
        print(lines[i])
    '''    
    metricRunTimePoint = [None]*runs
    metricDPErrorPoint = [None]*runs
    metricDPStoreErrorPoint = [None]*runs
    metricTTStoreErrorPoint = [None]*runs
    metricRunTimeRange = [None]*runs
    metricDPErrorRange = [None]*runs
    metricDPStoreErrorRange = [None]*runs
    metricTTStoreErrorRange = [None]*runs

    for i in range(runs):
        metricRunTimePoint[i] = string2vect(lines[(i*numPrint) + 1], 'metricRunTimePoint')
        metricDPErrorPoint[i] = string2vect(lines[(i*numPrint) + 2], 'metricDPErrorPoint')
        metricDPStoreErrorPoint[i] = string2vect(lines[(i*numPrint) + 3], 'metricDPStoreErrorPoint')
        metricTTStoreErrorPoint[i] = string2vect(lines[(i*numPrint) + 4], 'metricTTStoreErrorPoint')
        metricRunTimeRange[i] = string2vect(lines[(i*numPrint) + 5], 'metricRunTimeRange')
        metricDPErrorRange[i] = string2vect(lines[(i*numPrint) +6], 'metricDPErrorRange')
        metricDPStoreErrorRange[i] = string2vect(lines[(i*numPrint) + 7], 'metricDPStoreErrorRange')
        metricTTStoreErrorRange[i] = string2vect(lines[(i*numPrint) + 8], 'metricTTStoreErrorRange')
       

    metricRunTimePoint_mean = np.mean(np.array(metricRunTimePoint), axis = 0)
    metricDPErrorPoint_mean = np.mean(np.array(metricDPErrorPoint), axis = 0)
    metricDPStoreErrorPoint_mean = np.mean(np.array(metricDPStoreErrorPoint), axis = 0)
    metricTTStoreErrorPoint_mean = np.mean(np.array(metricTTStoreErrorPoint), axis = 0)
    metricRunTimeRange_mean = np.mean(np.array(metricRunTimeRange), axis = 0)
    metricDPErrorRange_mean = np.mean(np.array(metricDPErrorRange), axis = 0)
    metricDPStoreErrorRange_mean = np.mean(np.array(metricDPStoreErrorRange), axis = 0)
    metricTTStoreErrorRange_mean = np.mean(np.array(metricTTStoreErrorRange), axis = 0)
     
    d = dict(); 
    d['metricRunTimePoint_mean'] = metricRunTimePoint_mean
    d['metricDPErrorPoint_mean'] = metricDPErrorPoint_mean
    d['metricDPStoreErrorPoint_mean'] = metricDPStoreErrorPoint_mean
    d['metricTTStoreErrorPoint_mean'] = metricTTStoreErrorPoint_mean
    d['metricRunTimeRange_mean'] = metricRunTimeRange_mean
    d['metricDPErrorRange_mean'] = metricDPErrorRange_mean
    d['metricDPStoreErrorRange_mean'] = metricDPStoreErrorRange_mean
    d['metricTTStoreErrorRange_mean'] = metricTTStoreErrorRange_mean
    
    '''
    print(metricRunTimePoint)
    print(metricDPErrorPoint)
    print(metricDPStoreErrorPoint)
    print(metricTTStoreErrorPoint)
    print(metricRunTimeRange)
    print(metricDPErrorRange)
    print(metricDPStoreErrorRange)
    print(metricTTStoreErrorRange)
    '''
    
    return d
```

```python
def plotSort(T, arrayLeaf, arrayTree_baseline, arrayTree_sortD, xlable, ylable, title):
    xpoints = np.linspace(0, T, T, endpoint=True)
    y1points = arrayLeaf
    y2points = arrayTree_baseline
    y3points = arrayTree_sortD
    
    #plt.subplot(1, 1, 1)
    plt.plot(xpoints, y1points, 'b^', label="SortTime_leaf")
    plt.plot(xpoints, y2points, 'r^', label="SortTime_baseline")
    plt.plot(xpoints, y3points, 'g^', label="SortTime_tree_sortD")

    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.legend(loc="upper left", prop={'size': 8})

    plt.title(title)
    plt.show()
```

```python
def plotQ11(T, eps, N):
    # create file names
 
    T_string = str(T)+","
    eps_string = str(eps)+","
    N_string = str(N)
    fileNameLeaf = "leafbin40_ss1.txt,"+T_string+eps_string+N_string+";2.txt"
    fileNameBaseline = "baselinebin40_ss1.txt,"+T_string+eps_string+N_string+";2.txt"
    fileNameTreeSortD = "treebin40_ss1.txt,"+T_string+eps_string+N_string+",2;0.txt"

    # read files, parse and return mean
    returnedMeanLeaf = readFileReturnMeanAll(fileNameLeaf)
    returnedMeanBaseline = readFileReturnMeanBaseline(fileNameBaseline)
    returnedMeanSortD = readFileReturnMeanAll(fileNameTreeSortD)

    # three metrics
    #leaf 
    LeafRunTimePoint_mean = returnedMeanLeaf["metricRunTimePointQuery_mean"]
    LeafRunTimeRange_mean = returnedMeanLeaf["metricRunTimeRangeQuery_mean"]
    #baseline
    BaselineRunTimePoint_mean = returnedMeanBaseline["metricRunTimePoint_mean"]
    BaselineRunTimeRange_mean = returnedMeanBaseline["metricRunTimeRange_mean"]
    #tree_sortD
    SortDRunTimePoint_mean = returnedMeanSortD["metricRunTimePointQuery_mean"]
    SortDRunTimeRange_mean = returnedMeanSortD["metricRunTimeRangeQuery_mean"]
    
    title = "T:"+str(T)+"; eps:"+str(eps)+"; N:"+str(N)
    plotSort(T, LeafRunTimePoint_mean, BaselineRunTimePoint_mean, SortDRunTimePoint_mean, "time", "queryProcessing for point query(ms)", title)
    plotSort(T, LeafRunTimeRange_mean, BaselineRunTimeRange_mean, SortDRunTimeRange_mean, "time", "queryProcessing for range query(ms)", title)


    
```

```python
# Q1-runtime: EMP
# x axis: T
# y axis: query processing time for baseline and DP structure appoaches
T = 10
eps = 1
N = 10
plotQ11(T, eps, N)
 
```

```python
def plotDPSeperate(T, arrayLeaf, arrayTree_sortSubroot, arrayTree_sortD, xlable, ylable, title):
    xpoints = np.linspace(1, T, T, endpoint=True)
    y1points = arrayLeaf
    y2points = arrayTree_sortSubroot
    y3points = arrayTree_sortD

    #plt.subplot(1, 1, 1)
    plt.plot(xpoints, y1points, 'b^', label="leaf")
    plt.plot(xpoints, y2points, 'r^', label="baseline")
    plt.plot(xpoints, y3points, 'g^', label="tree_sortD")

    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.legend(loc="upper left", prop={'size': 8})

    plt.title(title)

    plt.show()
    
def plotQ12(T, eps, N):
    # create file names
    T_string = "T:"+str(T)
    eps_string = ",eps:"+str(eps)
    N_string = ",N:"+str(N)
    fileNameAll = "all-"+T_string+eps_string+N_string+".json"
    fileNameBaseline = "baseline-"+T_string+eps_string+N_string+".json"
    with open(fileNameAll) as data_file:
        data_loadedAll = json.load(data_file)
    with open(fileNameBaseline) as data_file:
        data_loadedBaseline = json.load(data_file)
    
    # six metrics 
    title = "T:"+str(T)+"; eps:"+str(eps)+"; N:"+str(N)
    # point query
    # leaf 
    LeafDPError_mean = data_loadedAll["list_metricDPError_leaf"]
    LeafDPStoreError_mean = data_loadedAll["list_metricDPStoreError_leaf"]
    LeafTTStoreError_mean = data_loadedAll["list_metricTTStoreError_leaf"]
    # baseline
    BaselineDPError_mean = data_loadedBaseline["list_metricDPErrorPoint_Baseline"]
    BaselineDPStoreError_mean = data_loadedBaseline["list_metricDPStoreErrorPoint_Baseline"]
    BaselineTTStoreError_mean = data_loadedBaseline["list_metricTTStoreErrorPoint_Baseline"]
    # tree_sortD
    SortDDPError_mean = data_loadedAll["list_metricDPError_treeD"]
    SortDDPStoreError_mean = data_loadedAll["list_metricDPStoreError_treeD"]
    SortDTTStoreError_mean = data_loadedAll["list_metricTTStoreError_treeD"]

    
    plotDPSeperate(T, LeafDPError_mean, BaselineDPError_mean, SortDDPError_mean, "time", "DPError-pointQ", title)
    plotDPSeperate(T, LeafTTStoreError_mean, BaselineTTStoreError_mean, SortDTTStoreError_mean, "time", "TrueStoreError-pointQ", title)
    plotDPSeperate(T, LeafDPStoreError_mean, BaselineDPStoreError_mean, SortDDPStoreError_mean, "time", "DPStoreError-pointQ", title)
 
    # range query
    # leaf 
    LeafDPErrorRange_mean = data_loadedAll["list_metricDPErrorRange_leaf"]
    LeafDPStoreErrorRange_mean = data_loadedAll["list_metricDPStoreErrorRange_leaf"]
    LeafTTStoreErrorRange_mean = data_loadedAll["list_metricTTStoreErrorRange_leaf"]
    # baseline
    BaselineDPErrorRange_mean = data_loadedBaseline["list_metricDPErrorRange_Baseline"]
    BaselineDPStoreErrorRange_mean = data_loadedBaseline["list_metricDPStoreErrorRange_Baseline"]
    BaselineTTStoreErrorRange_mean = data_loadedBaseline["list_metricTTStoreErrorRange_Baseline"]
    # tree_sortD
    SortDDPErrorRange_mean = data_loadedAll["list_metricDPErrorRange_treeD"]
    SortDDPStoreErrorRange_mean = data_loadedAll["list_metricDPStoreErrorRange_treeD"]
    SortDTTStoreErrorRange_mean = data_loadedAll["list_metricTTStoreErrorRange_treeD"]

    
    plotDPSeperate(T, LeafDPErrorRange_mean, BaselineDPErrorRange_mean, SortDDPErrorRange_mean, "time", "DPError-rangeQ", title)
    plotDPSeperate(T, LeafTTStoreErrorRange_mean, BaselineTTStoreErrorRange_mean, SortDTTStoreErrorRange_mean, "time", "TrueStoreError-rangeQ", title)
    plotDPSeperate(T, LeafDPStoreErrorRange_mean, BaselineDPStoreErrorRange_mean, SortDDPStoreErrorRange_mean, "time", "DPStoreError-rangeQ", title)


```

```python
def plotMulti(lst, leaf_lst, treeD_lst, treeA_lst, title):
    
    x_axis = np.arange(len(lst))
    # Multi bar Chart
    plt.bar(x_axis +0.20, leaf_lst, width=0.2, label = 'leaf')
    plt.bar(x_axis +0.20*2, treeD_lst, width=0.2, label = 'tree_sortD')
    plt.bar(x_axis +0.20*3, treeA_lst, width=0.2, label = 'tree_sortA')

    plt.xticks(x_axis,lst)
    plt.legend()
    plt.yscale('log',base=10) 
    plt.title(title)
    plt.show()
    
def plotQ3Point(p1, p2, p3, NorEps): #eps: T, N, epsList or N: T, eps, NList
    T = p1
    if NorEps == "eps":
        size = len(p3)
        epsList = p3
        N = p2
    else:
        size = len(p3)
        NList = p3
        eps = p2
        
    LeafRunTimeDP_list = [None]*size
    LeafRunTimeDPSort_list = [None]*size
    LeafRunTimeDPMerge_list = [None]*size
    LeafDPError_list = [None]*size
    LeafDPStoreError_list = [None]*size
    LeafTTStoreError_list = [None]*size
    
    SortRootRunTimeDP_list = [None]*size
    SortRootRunTimeDPSort_list = [None]*size
    SortRootRunTimeDPMerge_list = [None]*size
    SortRootDPError_list = [None]*size
    SortRootDPStoreError_list = [None]*size
    SortRootTTStoreError_list = [None]*size
    
    SortDRunTimeDP_list = [None]*size
    SortDRunTimeDPSort_list = [None]*size
    SortDRunTimeDPMerge_list = [None]*size
    SortDDPError_list = [None]*size
    SortDDPStoreError_list = [None]*size
    SortDTTStoreError_list = [None]*size
    # create file names
    for i in range(size):
        T_string = "all-T:"+str(T)
        if NorEps == "eps":
            eps_string = ",eps:"+str(p3[i])
            N_string = ",N:"+str(N)
        else:
            eps_string = ",eps:"+str(eps)
            N_string = ",N:"+str(p3[i])
        
        fileName = T_string+eps_string+N_string+".json"
        with open(fileName) as data_file:
            data_loaded = json.load(data_file)
         
        # six metrics
            #leaf 
        LeafRunTimeDPSort_list[i] = sum(data_loaded["list_runTimeDPSort_leaf"])
        LeafDPError_list[i] = sum(data_loaded["list_metricDPError_leaf"])
        LeafDPStoreError_list[i] = sum(data_loaded["list_metricDPStoreError_leaf"])
        LeafTTStoreError_list[i] = sum(data_loaded["list_metricTTStoreError_leaf"])
            #tree_sortRoot
        SortRootRunTimeDPSort_list[i] = sum(data_loaded["list_runTimeDPSort_treeA"])
        SortRootDPError_list[i] = sum(data_loaded["list_metricDPError_treeA"])
        SortRootDPStoreError_list[i] = sum(data_loaded["list_metricDPStoreError_treeA"])
        SortRootTTStoreError_list[i] = sum(data_loaded["list_metricTTStoreError_treeA"])
            #tree_sortD
        SortDRunTimeDPSort_list[i] = sum(data_loaded["list_runTimeDPSort_treeD"])
        SortDDPError_list[i] = sum(data_loaded["list_metricDPError_treeD"])
        SortDDPStoreError_list[i] = sum(data_loaded["list_metricDPStoreError_treeD"])
        SortDTTStoreError_list[i] = sum(data_loaded["list_metricTTStoreError_treeD"])
      #  print("SortRootRunTimeDPSort", returnedMeanSortRoot["metricRunTimeDPSort_mean"])
      #  print("SortDRunTimeDPSort_list", returnedMeanSortD["metricRunTimeDPSort_mean"])
      #  print("SortRootRunTimeDPSort", sum(returnedMeanSortRoot["metricRunTimeDPSort_mean"]))
      #  print("SortDRunTimeDPSort_list", sum(returnedMeanSortD["metricRunTimeDPSort_mean"]))

        
    xList = [str(x) for x in p3]
    
    #sortTime
    plotMulti(xList, LeafRunTimeDPSort_list, SortDRunTimeDPSort_list, SortRootRunTimeDPSort_list, "SortTime--varying "+NorEps)

    # accuracy 
    plotMulti(xList, LeafDPError_list, SortDDPError_list, SortRootDPError_list, "PointQ-DPCountError--varying "+NorEps)
    plotMulti(xList, LeafDPStoreError_list, SortDDPStoreError_list, SortRootDPStoreError_list, "PointQ-DPStoreError--varying "+NorEps)
    plotMulti(xList, LeafTTStoreError_list, SortDTTStoreError_list, SortRootTTStoreError_list, "PointQ-TTStoreError--varying "+NorEps)
    
    
def plotQ3Range(p1, p2, p3, NorEps): #eps: T, N, epsList or N: T, eps, NList
    T = p1
    if NorEps == "eps":
        size = len(p3)
        epsList = p3
        N = p2
    else:
        size = len(p3)
        NList = p3
        eps = p2
        
    LeafRunTimeDP_list = [None]*size
    LeafRunTimeDPSort_list = [None]*size
    LeafRunTimeDPMerge_list = [None]*size
    LeafDPError_list = [None]*size
    LeafDPStoreError_list = [None]*size
    LeafTTStoreError_list = [None]*size
    
    SortRootRunTimeDP_list = [None]*size
    SortRootRunTimeDPSort_list = [None]*size
    SortRootRunTimeDPMerge_list = [None]*size
    SortRootDPError_list = [None]*size
    SortRootDPStoreError_list = [None]*size
    SortRootTTStoreError_list = [None]*size
    
    SortDRunTimeDP_list = [None]*size
    SortDRunTimeDPSort_list = [None]*size
    SortDRunTimeDPMerge_list = [None]*size
    SortDDPError_list = [None]*size
    SortDDPStoreError_list = [None]*size
    SortDTTStoreError_list = [None]*size
    # create file names
    for i in range(size):
        T_string = "all-T:"+str(T)
        if NorEps == "eps":
            eps_string = ",eps:"+str(p3[i])
            N_string = ",N:"+str(N)
        else:
            eps_string = ",eps:"+str(eps)
            N_string = ",N:"+str(p3[i])
        
        fileName = T_string+eps_string+N_string+".json"
        with open(fileName) as data_file:
            data_loaded = json.load(data_file)
         
        # six metrics
            #leaf 
        LeafRunTimeDPSort_list[i] = sum(data_loaded["list_runTimeDPSort_leaf"])
        LeafDPError_list[i] = sum(data_loaded["list_metricDPErrorRange_leaf"])
        LeafDPStoreError_list[i] = sum(data_loaded["list_metricDPStoreErrorRange_leaf"])
        LeafTTStoreError_list[i] = sum(data_loaded["list_metricTTStoreErrorRange_leaf"])
            #tree_sortRoot
        SortRootRunTimeDPSort_list[i] = sum(data_loaded["list_runTimeDPSort_treeA"])
        SortRootDPError_list[i] = sum(data_loaded["list_metricDPErrorRange_treeA"])
        SortRootDPStoreError_list[i] = sum(data_loaded["list_metricDPStoreErrorRange_treeA"])
        SortRootTTStoreError_list[i] = sum(data_loaded["list_metricTTStoreErrorRange_treeA"])
            #tree_sortD
        SortDRunTimeDPSort_list[i] = sum(data_loaded["list_runTimeDPSort_treeD"])
        SortDDPError_list[i] = sum(data_loaded["list_metricDPErrorRange_treeD"])
        SortDDPStoreError_list[i] = sum(data_loaded["list_metricDPStoreErrorRange_treeD"])
        SortDTTStoreError_list[i] = sum(data_loaded["list_metricTTStoreErrorRange_treeD"])
      #  print("SortRootRunTimeDPSort", returnedMeanSortRoot["metricRunTimeDPSort_mean"])
      #  print("SortDRunTimeDPSort_list", returnedMeanSortD["metricRunTimeDPSort_mean"])
      #  print("SortRootRunTimeDPSort", sum(returnedMeanSortRoot["metricRunTimeDPSort_mean"]))
      #  print("SortDRunTimeDPSort_list", sum(returnedMeanSortD["metricRunTimeDPSort_mean"]))

        
    xList = [str(x) for x in p3]
    
    #sortTime
    plotMulti(xList, LeafRunTimeDPSort_list, SortDRunTimeDPSort_list, SortRootRunTimeDPSort_list, "SortTime--varying "+NorEps)

    # accuracy 
    plotMulti(xList, LeafDPError_list, SortDDPError_list, SortRootDPError_list, "RangeQ-DPCountError--varying "+NorEps)
    plotMulti(xList, LeafDPStoreError_list, SortDDPStoreError_list, SortRootDPStoreError_list, "RangeQ-DPStoreError--varying "+NorEps)
    plotMulti(xList, LeafTTStoreError_list, SortDTTStoreError_list, SortRootTTStoreError_list, "RangeQ-TTStoreError--varying "+NorEps)
```
