```python
import math
import matplotlib.pyplot as plt
import numpy as np
import re
import json
```

```python
with open("DP-T:200,eps:1,N:1000.json") as data_file:
    data_loaded = json.load(data_file)
    LeafDPError_mean = data_loaded["list_metricDPError_leaf"]
  
    SortRootDPError_mean = data_loaded["list_metricDPError_treeA"]
 
    SortDDPError_mean = data_loaded["list_metricDPError_treeD"]
    T = 200
    x1 = np.linspace(1, T, T)
    log2 = [math.log2(i+1) for i in range(T)]
    plt.plot(x1, LeafDPError_mean, 'b^', label="Leaf ")
    plt.plot(x1, SortRootDPError_mean, 'r^', label="Tree-baseline ")
    plt.plot(x1, SortDDPError_mean, 'g^', label="Tree-optimized ")

    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.legend(loc="upper right", prop={'size': 10})
    plt.yscale('log',base=10) 
    
    plt.show()
```
